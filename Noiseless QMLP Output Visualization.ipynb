```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file
df = pd.read_csv("MNIST_Amplitude.csv")

# Define plot style
# Define plot style
plt.figure(figsize=(10, 6))
markers = {
    'LF': 'x',    # without label smoothing
    'LFLS': 'o'   # with label smoothing
}
colors = {
    2: 'blue',
    10: 'green',
    50: 'red'
}

# Plot each column
for col in df.columns[1:]:  # Skip 'Epoch'
    parts = col.strip().split()  # e.g., ['L', '2', 'LFLS']
    if len(parts) == 3:
        layer = int(parts[1])
        ls_type = parts[2]
    else:
        continue  # Skip columns with unexpected format

    label = f"Layer {layer} - {ls_type}"
    plt.plot(df['Epoch'].to_numpy(), df[col].to_numpy(),
             marker=markers[ls_type], linestyle='-', color=colors[layer], label=label)

# Formatting with larger font sizes
plt.xlabel("Epoch", fontsize=24)
plt.ylabel("Accuracy", fontsize=24)
plt.title("Accuracy vs Epoch QMLP (MNIST Amplitude Encoding)", fontsize=22)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.ylim(0, 100)
plt.grid(True)
plt.legend(fontsize=20)
plt.tight_layout()

# Show plot
plt.show()
```
