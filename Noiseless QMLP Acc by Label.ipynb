```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load CSV file
df = pd.read_csv("MNIST_Amplitude acc.csv")

# Get the last row (final accuracies) or average accuracy â€” pick one
# Here we use the last row
final_acc = df.iloc[-1]

# Clean up column names
final_acc.index = final_acc.index.str.strip()

# Organize data by layer
layers = [2, 10, 50]
categories = ['Clean', 'WLS', 'LS']

# Create a data structure: rows=layers, cols=Clean/WLS/LS
data = []
for layer in layers:
    row = [
        final_acc[f'L{layer} Clean'] if f'L{layer} Clean' in final_acc else final_acc[f'L {layer} Clean'],
        final_acc[f'L {layer} WLS'],
        final_acc[f'L {layer} LS']
    ]
    data.append(row)

data = np.array(data)

# Plotting
x = np.arange(len(layers))  # Layer positions
bar_width = 0.25

plt.figure(figsize=(8, 5))

# Plot bars
for i, category in enumerate(categories):
    plt.bar(x + i * bar_width, data[:, i], width=bar_width, label=category)

# Formatting
plt.xticks(x + bar_width, [f"Layer {l}" for l in layers], fontsize=24)
plt.yticks(fontsize=24)
plt.ylabel("Accuracy", fontsize=24)
plt.title("MNIST Amplitude Encoding", fontsize=24)
plt.legend(fontsize=22)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

# Show plot
plt.show()
```
