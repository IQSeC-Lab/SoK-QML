```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file
df = pd.read_csv("MNIST_Amplitude.csv")

# Define plot style
plt.figure(figsize=(10, 6))
markers = {
    'WLS': 'x',  # without label smoothing
    'LS': 'o'    # with label smoothing
}
colors = {
    2: 'blue',
    10: 'green',
    50: 'red'
}

# Plot each column
for col in df.columns[1:]:  
    parts = col.split() 
    layer = int(parts[1])
    ls_type = parts[2]
    label = f"Layer {layer} - {ls_type}"
    plt.plot(df['Epoch'].to_numpy(), df[col].to_numpy(),
             marker=markers[ls_type], linestyle='-', color=colors[layer], label=label)

# Formatting with larger font sizes
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Accuracy", fontsize=14)
plt.title("Accuracy vs Epoch QMLP (MNIST Amplitude Encoding)", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)
plt.legend(fontsize=14)
plt.tight_layout()

# Show plot
plt.show()
```
